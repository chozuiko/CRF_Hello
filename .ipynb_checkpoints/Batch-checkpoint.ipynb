```python
from bs4 import BeautifulSoup
import requests
import json
import pandas as pd
from py2neo import authenticate, Graph

USERNAME = "neo4j"  # use your actual username
PASSWORD = "iwate"  # use your actual password
authenticate("localhost:7474", USERNAME, PASSWORD)
graph = Graph()

#merge top level concept
query_start = """
merge (top:Entity{name:{toplevel}})
WITH top, {list} AS entities
UNWIND entities AS entity
MERGE (en:Entity {name:entity})-[:is_a]->(top)
"""
#merge all concept
query_all = '''
WITH {list} AS all
unwind all as entity
merge (en:Entity{name:entity})
'''
#add is_a relationships
query_is_a = '''
match (par:Entity{name:{parent}})
with par,{children} as children
unwind children as child
match (chi:Entity{name:child})
merge (chi)-[:is_a]->(par)
'''

df = pd.read_csv('./extracted.csv', names=['title', 'url', 'heads', 'bolds'])
df['heads'] = df['heads'].astype(object)
df['bolds'] = df['bolds'].astype(object)


for index, row in df.iterrows():
    url = str(row["url"])
    html_string = requests.get(url).text
    soup = BeautifulSoup(str( html_string ),  'lxml')
    try:
        top_level = soup.find("a", class_="sidebar_component__title__title").get_text()
        
        ul = soup.find("ul", class_="related_links__sub_section")
        start = [li.get_text() for li in ul.find_all("li", recursive=False)]
        all = [li.get_text() for li in ul.find_all("li")]
    except:
        top_level = []
        start = []
        all = []
    graph.run(query_start, list=start, toplevel=top_level)
    graph.run(query_all, list=all)
    try:
        for temp in ul.find_all("ul"):
            if temp.name == "ul":
                parent = temp.previousSibling.get_text()
                print(parent)
                children = [li.get_text() for li in temp.find_all("li", recursive=False)]
                print(children)
                graph.run(query_is_a, parent=parent, children=children)
    except:
        pass
    print("sucess") 
       





```
